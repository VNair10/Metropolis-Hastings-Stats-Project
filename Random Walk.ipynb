```python
#random walk
import numpy as np
import matplotlib.pyplot as plt
import random
N = 1000
random.seed(12564)
a = np.zeros(N)
b = np.zeros(N)
for i in range(1, N):
    v = random.randint(1, 9)
    if v == 1:
        a[i] = a[i - 1] + 1
        b[i] = b[i - 1]
    elif v==2:
        a[i] = a[i - 1]+ 1
        b[i] = b[i - 1] - 1
    elif v==3:
        a[i] = a[i - 1]
        b[i] = b[i - 1] - 1
    elif v==4:
        a[i] = a[i - 1]- 1
        b[i] = b[i - 1] - 1
    elif v == 5:
        a[i] = a[i - 1] - 1
        b[i] = b[i - 1]
    elif v==6:
        a[i] = a[i - 1]- 1
        b[i] = b[i - 1] + 1
    elif v == 7:
        a[i] = a[i - 1]
        b[i] = b[i - 1] + 1
    elif v==8:
        a[i] = a[i - 1]+ 1
        b[i] = b[i - 1] + 1
    else:
        a[i] = a[i - 1]
        b[i] = b[i - 1]
    
    
plt.title("Random Walk N=%s "%N)
plt.plot(a,b)
```
