```python
#monte carlo integration

from tabulate import tabulate
import numpy as np
np.random.seed(1693)
np.set_printoptions(suppress=True)
N=np.power(10,[3,4,5,6,7])
e=[]
a = 0
b = 1
for i in N:
  x = np.random.uniform(a,b,i)
  fx = x**2
  e.append(np.mean(fx)*(b-a))
print(tabulate(np.array([N,e]).T,headers=['N','I1']))
e=[]
for i in N:
  x3 = np.random.uniform(0, 1,i)
  x2 = np.random.uniform(0,1-x3,i)
  x1 = np.random.uniform(0,1-x2-x3,i)
  gx = np.exp(x1**(x2**2+x3**2)+x2**(x3**2)+x3**2)
  e.append(np.mean(gx*(1-x2-x3)*(1-x3)))
print(tabulate(np.array([N,e]).T,headers=['N','I3']))
```
